```python
# Հիմնական դաս
class Animal:
    def __init__(self, name): # function -> method
        self.name = name
        
	def get_name(self):
        print(self.name)
        
    def make_sound(self):
        pass  # Այս մեթոդը կվերասահմանեն ժառանգները

# Ժառանգություն
class Dog(Animal):
    def make_sound(self):
        return "Բարև, ես շուն եմ, վաֆ-վաֆ!"

class Cat(Animal):
    def make_sound(self):
        return "Բարև, ես կատու եմ, մյաու-մյաու!"
```

```python
# Հատուկ մեթոդներ (Magic Methods)
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def __str__(self):  # Հատուկ մեթոդ տպելու համար
        return f"Անուն: {self.name}, Տարիք: {self.age}"
    
    def __add__(self, other):  # Հատուկ մեթոդ գումարման համար
        return self.age + other.age

# isinstance և issubclass
print(isinstance(Dog("Rex"), Animal))
print(issubclass(Dog, Animal))
```

```python
# Բազմակի ժառանգականություն
class Flyable:
    def fly(self):
        return "Ես կարող եմ թռչել!"

class Bird(Animal, Flyable):
    def make_sound(self):
        return "Չիվ-չիվ!"
```
